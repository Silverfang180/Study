- Variable name should not start with Numbers(0 to 9), but we can use number in between & at the end of variable.
- In Variable name special characters(!@#$%^&*) are not allowed
- Variables are case sensitive

# Lists
- List is an ordered collections of hetrogeneous data type values
- List follows indexing and slicing
- List is mutable in nature. It means we can add new element, modifiy existing element and deleting existing element.
- List elements enclosed in **[]** and separated by commas  (**,**)



```python
lst = ['AFK',34,3534.45]
a=lst[2]
b=lst[-1]
print(a)
print(b)
```

# Index:- Indexing is a way to retrive perticular way.
## Positive Indexing
- It starts from 0  goes upto n-1.
- it travels from left to right.
## Negative Indexing
- It starts from -1 and goes upto n-1.
- It travels from right to left.
# Slicing
- It is a way to retrive sub-sequence
- working:-
[start->0, stop-> n-1, step-> +1]

```python
#Positive indexing
#      0     1   2     3    4
lst=["hello",23,44.43,True,None]
```

```python
#Negative indexing
#     -6     -5   -3     -2    -1
lst=["hello", 23, 44.43, True, None]
```

```python
# eg. of slicing
lst=[1,2,1,4,5,65,45,12]
a=lst[0:5]
b=lst[-8:-3]
print(a,b)
```

```python
lst = [2,7,76,23,54,77,90,21,6,9,10] # [54,23,76,7]
lst[4:0:-1]
```

##Mutability
- add :- it uses append,extend, insert functions , it adds single elments at the end of the list
- modify :-
- delete

```python
#Append
lst = [2,4,55]
lst.append('Abhib')
print(lst)
```

```python
#extends : It adds seq at the end of the list
lst2 = [5,6,7]
lst3 = ['A', 'b']
lst2.extend(lst3)
print(lst2)
```

```python
# Insert - it add single elements as specific index of the list(Indexing, Value)

lst= [2,41,26]
lst.insert(2,'Abhib')
print(lst)
```

```python
# Modifing :- In this operation we modify the exisitng elements

lst = [2,34,5]
lst[1] = 10 # here the elements on index 1 modified to 10 from 34
lst
```

```python
# Delete

lst = [4,5,67,8]
lst.pop() # In pop() it by deflaut last element
lst
```

```python
lst = [4,5,67,8]
lst.pop(2) # In pop() with the help of indexing it delete that particular value
lst
```

```python
# Remove :- it removes specific values present in list

lst = [3,4,5]
lst.remove(4)
lst
```

```python
# Clear :- it deletes/clears all the data/values present in list

lst = [4,5,6]
lst.clear()
lst
```
