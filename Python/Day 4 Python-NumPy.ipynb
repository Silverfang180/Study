#**Numpy Array**
- length of numpy array is fixed
- Homogeneous elements are stored
- Advanced mathematical operation can be perforemd using built in functions
- It supports advanced slicing and indexing
- Faster as compared to python List


#**central of tendency**
- Mean
- Median
- Mode

```python
#Mean
data = np.array([10, 20, 3, 40, 50])
np.mean(data)
```

```python
#Mode : Numpy has no attributes called mode , so we use Scipy when we want mode of any given array
sp.stats.mode(data)
```

```python
#Median
np.median(data)
```

#**Measure of Dispersion**
- Variance
- Range
- Standard Deviation

```python
# Variance
data = np.array([10, 20, 3, 40, 50])
np.var(data)
```

```python
# Range
np.ptp(data)
```

```python
# Standard Deviation
np.std(data)
```

```python
data = np.array([10, 20, 3, 40, 50])
np.quantile(data,0.0)
```

```python
np.quantile(data,0.25)
```
